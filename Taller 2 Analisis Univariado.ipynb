Introducción.

Se va a realizar el analisis de un conjunto de datos con información de varibales usadas para predicción de cancer de Seno. Este conjunto de datos es de la libreria de SKLearn.

La variable a analizar va a ser mean compactness.

Se selecciona esta variable porque creemos que puede aportar información importante relacionada con el cancer de seno porque el cancer de seno normalmente se puede iniciar a detectar con autoexamenes que consisten en deteccion de masas en el seno, entonces si hay masas creemos que el nivel de compatisidad del seno va a ser mayor y puede arrojar información relevante en cuanto a la detección de esta enfermedad.


Código en Python con análisis univariado y visualizaciones

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df.columns
```

```python
var=df["mean compactness"]
var.describe()
```

```python
var.isnull().sum()
```

```python
sns.histplot(var,kde=True)
plt.show()
```

```python
sns.boxplot(var)
plt.show()
```

Interpretación de los resultados

Podemos observar que no tenemos datos nulos y que contamos con un total de 569 registros los cuales se encuentran en un rango de [0.01938 , 0.3454].
 En el histograma vemos que la distribución no es normal, esta es asimetrica a la derecha en la que se ve una cola al lado derecho del histograma.
 En el boxplot de reafirma el sesgo que tene la información y se puede ver que hay algunos outliers.
